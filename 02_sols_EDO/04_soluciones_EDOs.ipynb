### Ejemplo 1: Método de Euler

Resolvamos la ecuación
$$y' = y - t^2 + 1$$

con condición inicial $$y(0) = 0.5$$

```python
from math import ceil # usaremos la función ceil más adelante
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h = 0.1 # Paso de tiempo
t = 0 
y = 0.5 

print(' Al paso 0: t = %1.2f, y = %1.2f\n' % (t, y))
```

```python
# Arrgelos para guardar las variables integradas (t, y)
TE1 = np.empty(50)
YE1 = np.empty(50)

for ii in range(50): 
    k1 = h * (y-2*t**2+1) # h*f(t,y)
    y = y + k1
    t = t + h
    TE1[ii] = t
    YE1[ii] = y
    print(' Al paso %d: t = %6.3f, y = %18.15f\n' % (ii,t, y))
```

```python
# Solución analítica a la ecuación 

YA1 = 2*TE1 - np.exp(TE1)/2 + TE1**2 + 1
```

```python
fig,(ax0,ax1) = plt.subplots(1,2,figsize=(12,4))

ax0.plot(TE1,YE1,'-ob', label='Euler')
ax0.plot(TE1,YA1,'-*m',label='Analítica')
ax0.set_xlabel('t')
ax0.set_ylabel('y')
ax0.legend()

ax1.stem(TE1,YA1-YE1,'b',use_line_collection=True)
ax1.set_xlabel('t')
ax1.set_ylabel(r'$y_a-y_e$')
ax1.set_title('Diferencia entre Euler y solución analítica')
```

### Ejemplo 2 - Método de Euler

Ahora resolvamos la ecuación
$$y'= sin(t)$$
con la condición inicial
$$y(0) = -1$$

y 
$$y = -cos(t)$$

```python
h=0.01;  # este es el incremento de tiempo 
t = 0;   # tiempo inicial, t0
y = -1;  # valor de y en t0

print('Paso 0: t = %6.3f, y = %18.15f\n' %(t, y))
```

```python
TE2 = np.empty(ceil(4*np.pi/h))
YE2 = np.empty(ceil(4*np.pi/h))

for ii in range(ceil(4*np.pi/h)):
    k1 = h * np.sin(t)                   # h*f(t,y)
    y = y + k1
    t = t + h
    TE2[ii] = t
    YE2[ii] = y    
    print('Paso %d: t = %6.3f, y = %18.15f\n' % (ii, t, y))
```

```python
# Solución analítica    
YA2 = -np.cos(TE2);
```

```python
fig,(ax0,ax1) = plt.subplots(1,2,figsize=(12,4))

ax0.plot(TE2,YE2,'-og', label='Euler')
ax0.plot(TE2,YA2,'-*m',label='Analítica')
ax0.set_xlabel('t')
ax0.set_ylabel('y')
ax0.legend()

ax1.stem(TE2,YA2-YE2,'b',use_line_collection=True)
ax1.set_xlabel('t')
ax1.set_ylabel(r'$y_a-y_e$')
ax1.set_title('Diferencia entre Euler y solución analítica')
```

### Ejemplo 1 Método de Runge-Kutta orden 2 

Ahora resolvamos la misma ecuación que en el ejemplo 1 del método de Euler pero usando el método de Runge-Kutta: $$ y'=y - t^2 + 1$$ con $$y(0) = 0.5$$

Recuerden que Runge Kutta orden 1 es simplemente Euler y Runge Kutta orden 2 es el método del punto medio. ¿Pueden probarlo?

```python
h = 0.01 # Paso de tiempo
t = 0
y = 0.5 
print('Paso 0: t = %6.3f, y = %18.15f\n' %(t, y))

TRK1 = np.empty(int(5/h))
YRK1 = np.empty(int(5/h))

for ii in range(int(5/h)):
    k1 = h * (y - t**2 + 1)              # h*f(t,y);
    k2 = h * ((y+k1) - (t+h)**2 + 1) # h*f(t+h, y+k1);
    y = y + (k1 + k2)/2
    t = t + h
    TRK1[ii] = t
    YRK1[ii] = y    
    print('Paso %d: t = %6.3f, y = %18.15f\n' % (ii, t, y))
```

```python
# Sol. analítica
YARK1 = 2*TRK1 - np.exp(TRK1)/2 + TRK1**2 + 1 
```

```python
fig,(ax0,ax1) = plt.subplots(1,2,figsize=(12,4))

ax0.plot(TRK1,YRK1,'-og', label='Runge-Kutta 2')
ax0.plot(TRK1,YARK1,'-*m',label='Analítica')
ax0.set_xlabel('t')
ax0.set_ylabel('y')
ax0.legend()

ax1.stem(TRK1,YARK1-YRK1,'b',use_line_collection=True)
ax1.set_xlabel('t')
ax1.set_ylabel(r'$y_a-y_{rk}$')
ax1.set_title('Diferencia entre RK 2 y solución analítica') 
```

¿Cómo se compara el error de la solución numérica usando el método de Euler y el de Runge-Kutta 2 para la EDO del ejemplo 1?

* Ambos errores aumentan en el tiempo, es decir, ambas soluciones numéricas se alejan cada vez más de la solución analítica.
* El error del método de Euler después de t=5 es 200 mientras que el de Runge-Kutta 2 es aproximadamente 0.0012, es decir, el error de RK2 después de t=5 ¡es unas 150 mil veces más pequeño!


### Ejemplo 2 Runge-Kutta

Ahora resolvamos la misma ecuación que en el ejemplo 2 del método de Euler usando Runge-Kutta 2:
    $$y'= sin(t)$$
    $$y(0) = -1.$$

```python
h=0.01;   # este es el incremento de tiempo o del_t
t = 0;   # tiempo inicial o t0
y = -1;  # valor de y en t0
print('Paso 0: t = %6.3f, y = %18.15f\n' %(t, y))
```

```python
def RK_f(a,b):
    return(np.sin(a))
```

```python
TRK2 = np.empty(ceil(4*np.pi/h))
YRK2 = np.empty(ceil(4*np.pi/h))

for ii in range(ceil(4*np.pi/h)):
    k1 = h * RK_f(t, y)              # h*f(t,y);
    k2 = h * RK_f(t+h, y+k1)     # h*f(t+h, y+k1);
    y = y + (k1 + k2)/2
    t = t + h
    TRK2[ii] = t
    YRK2[ii] = y
    print('Paso %d: t = %6.3f, y = %18.15f\n' %(ii, t, y))
```

```python
YARK2 = -np.cos(TRK2) # solución analítica
```

```python
fig,(ax0,ax1) = plt.subplots(1,2,figsize=(12,4))

ax0.plot(TRK2,YRK2,'-og', label='Runge-Kutta 2')
ax0.plot(TRK2,YRK2,'-*m',label='Analítica')
ax0.set_xlabel('t')
ax0.set_ylabel('y')
ax0.legend()

ax1.stem(TRK2,YARK2-YRK2,'b',use_line_collection=True)
ax1.set_xlabel('t')
ax1.set_ylabel(r'$y_a-y_{rk}$')
ax1.set_title('Diferencia entre RK2 y solución analítica')
```

De nuevo, comparemos el error del método de Runge-Kutta 2 y Euler para la ecuación del ejemplo 2. 

Ambos errores oscilan en el tiempo pero el error máximo (en valor abosluto) es muchísimo mayor para el método de Euler.
